```python
token = Tokenizer()
token.fit_on_texts([corpus])
total_words = len(token.word_index) + 1

dataset = []
for i in corpus.split("."):
    input_seq = token.texts_to_sequences([i])[0]
    for j in range(1, len(input_seq)):
        ngram = input_seq[:j + 1]
        dataset.append(ngram)
```

```python
from keras import utils
maxi = max([len(i) for i in dataset])
padded = pad_sequences(dataset, maxlen=maxi)

X = np.array([i[:-1] for i in padded])
Y = [i[-1] for i in padded]

y = np.array(utils.to_categorical(Y, num_classes=total_words))
```

```python
from keras.layers import Embedding, LSTM, Dense
model = keras.Sequential([
    Embedding(total_words, 100, input_length=maxi - 1),
    LSTM(150),
    Dense(total_words, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X, y, epochs=100, verbose=1)
```
